```python
# Turn an image into pencil sketch
```

```python
def dodge(front,back):
    result=front*255/(255-back) 
    result[result>255]=255
    result[back==255]=255
    return result.astype('uint8')
```

```python
import numpy as np
def grayscale(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
# Photo of Mashrafe Mortaza from ESPNcricinfo
```

```python
img ="http://p.imgci.com/db/PICTURES/CMS/263500/263576.jpg"
```

```python
import imageio
s = imageio.imread(img)
g=grayscale(s)
i = 255-g
import scipy.ndimage
b = scipy.ndimage.filters.gaussian_filter(i,sigma=10)
r= dodge(b,g)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.title('Mashrafe Mortaza', fontsize=16)
plt.imshow(r, cmap="gray")
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
plt.imsave('mashrafe_sketch.png', r, cmap='gray', vmin=0, vmax=255)
```
